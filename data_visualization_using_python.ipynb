# MATPLOTLIB

* Open source drawing library which supports rich drawing types
* Used to draw 2D and 3D graphics

---

#### Importing the module:

```python
import matplotlib.pyplot as plt
```

```python
x = [2,3,5,6,8]
y = [21,31,45,67,78]
plt.plot(x,y)
```

```python
sale_qty = [1200, 1430, 2300, 1500, 1700]
sale_month = ["Jan", "Feb", "March", "April", "May"]
plt.plot(sale_month, sale_qty)
```

```python
# adding labels to the axes and graph
plt.plot(sale_month, sale_qty)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales vs Month Graph")
```

```python
# changing the colour
plt.plot(sale_month, sale_qty, color="#AD0DE2")
```

```python
# adding points
plt.plot(sale_month, sale_qty, color="#F38989", marker = "*")
```

```python
# changing linestyle
plt.plot(sale_month, sale_qty, color = "k", marker = "_", linestyle = "dashed") #"dashed" or "--"
```

```python
# changing linestyle
plt.plot(sale_month, sale_qty, color = "y", marker = "_", linestyle = "dotted") #"dotted" or ":"
```

```python
# changing the linewidth
plt.plot(sale_month, sale_qty,
         color = "m",
         marker = "2",
         markersize = 24,
         linestyle = "--",
         linewidth = 7)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales vs Month Graph")
```

#### Two Graphs Comparision:

```python
sale1 = [200,350,250,380,400]
sale2 = [170,380,280,350,420]
month = ["Jan", "Feb", "Mar", "Apr", "May"]
plt.plot(month, sale1, marker="o", color="y", linestyle=":")
plt.plot(month, sale2, marker="2", color="k", linestyle="-.", linewidth=5)

plt.xlabel("Months")
plt.ylabel("some numbers")
plt.title("something vs Month Graph")
```

```python
sale1 = [200,350,250,380,400]
sale2 = [170,380,280,350,420]
month = ["Jan", "Feb", "Mar", "Apr", "May"]
plt.plot(month, sale1, marker="o", color="y", linestyle=":", label="sales1")
plt.plot(month, sale2, marker="2", color="k", linestyle="-.", linewidth=5, label="sales2")

plt.xlabel("Months")
plt.ylabel("some numbers")
plt.title("something vs Month Graph")

plt.legend(loc=4) #location of legend
```

# Bar Plot

```python
Month = ["Jan", "Feb", "Mar", "Apr", "May"]
Sale_Qty = [100,200,150,300,250]
plt.bar(Month, Sale_Qty, color="k")
plt.xlabel("Months")
plt.ylabel("Sale_Qty")
plt.title("Bar Plot")
```

#### Horizontal Bar Plots:

```python
Month = ["Jan", "Feb", "Mar", "Apr", "May"]
Sale_Qty = [100,200,150,300,250]
plt.barh(Month, Sale_Qty, color="r")
plt.xlabel("Months")
plt.ylabel("Sale_Qty")
plt.title("Bar Plot")
```

# Histogram:
* A histogram is a graph showing frequency distribution.
* It is a graph showing the bumber of obersvations within each given interval.

```python
ages = [11,23,56,42,23,25,84,45,65,75,43,52,84,76,82,34]
plt.hist(ages, bins = 20, color="c", edgecolor="k") #bins = no. of bars
plt.xlabel("Ages")
plt.title("Histogram Graph")
```

```python
import numpy as np
x1 = np.random.randn(50) #randn = normally distributed numbers of mean=0 and variance=1
x1
```

```python
plt.hist(x1, bins = 10, color="m", edgecolor="k", linestyle="-.")
plt.xlabel("Intervals")
plt.ylabel("Frequency")
plt.title("Random Hist")
```

# Pie Chart:

```python
qty = [25,15,20,30,30]
city = ['Pune', 'Nashik', 'Kolhapur', 'Mumbai', 'Nagpur']
plt.pie(qty, labels = city, startangle=35)
plt.show()
```

```python
activity = ['Sleep', 'Eat', 'Study', 'Work', 'Play', 'Entertainment']
hrs_spent = [7,2,5,3,3,4]
plt.pie(hrs_spent, labels = activity)
plt.show()
```

```python
activity = ['Sleep', 'Eat', 'Study', 'Work', 'Play', 'Entertainment']
hrs_spent = [7,2,5,3,3,4]
plt.pie(hrs_spent,
        labels = activity,
        startangle = 90,
        shadow = True,
        #colors=["w","c","w","c","w","c"],
        explode = (0,0,0.1,0,0,0.3), #explode seperates that percent from the whole graph.
        autopct = "%1.1f%%") #shows the percentage of the activities mentioned.
plt.title("Activities Pie") 
plt.show()
```

# Scatter Plot:

```python
import pandas as pd
```

```python
appt_price = {
    "area_sqft" : [350,400,450,500,540,600,660,700,780,800,900,910,990,1100,1200],
    "price_lacs" : [21.2, 23, 24, 26, 26.7, 29, 31.5, 32, 32.5, 35, 40, 45, 52, 61, 66]
}

df = pd.DataFrame(appt_price)
df
```

```python
plt.figure(figsize = (10,6))
plt.scatter(df["area_sqft"],df["price_lacs"], marker="2", color="r")

plt.xlabel("Area in sqft.")
plt.ylabel("Price in Lacs")
plt.title("Apt. Prices vs Area")
plt.show()
```
